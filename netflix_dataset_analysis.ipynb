```python
# * Manipulating the index *
#release year will be put in the beginning
netflix.set_index("release_year")
#If we want to check a certain show is available on netflix
netflix.loc[netflix.title.isin(["Kota Factory"])]
#If we want to check if any data is empty in our dataset(gives boolean output)
netflix.isnull()
```

```python
# * Assigning a data *
#Here we are adding a new column called certified and giving it a default value not certified which we will change according to our conditions 
netflix["certified"]=" not certified"
netflix["rating"]=netflix1["rating"]
netflix.head(10)
```

```python
#Summary Functions and Maps
#For example we want to know the number of shows of a certain rating type
netflix.rating.value_counts()
#This function is to get the names of the different types of ratings that are available
netflix.rating.unique()
#MAPS
#If we want to change the status of all shows from not certified to certified
def remark_ratings(row):
    row.certified="certified"
    return row
netflix.apply(remark_ratings,axis="columns")    
```

```python
#Now if we want to find the oldest show available in this Netflix data
oldest_show=netflix.release_year.idxmin()
oldest_show_is=netflix.loc[oldest_show,"title"]
print("The oldest show is ",oldest_show_is)#Here we have managed to find the oldest show in the given netflix data Pioneers: First Women Filmmakers 
```

```python
#If we want to find a particular genre or type of show
#Let us consider we want to know the number of documentaries available
n_doc=netflix.listed_in.map(lambda desc:"Documentaries" in desc).sum()
documentary_counts=pd.Series([n_doc],index=["Documentaries"])
print(documentary_counts)
```

```python
#Now if we want a list of number of  movies and shows by a certain director
netflix.groupby("director").size()
```
